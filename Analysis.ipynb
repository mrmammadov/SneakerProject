```python
sneakers = pd.read_excel(r'C:\Users\z0044zfu\Desktop\Topic 7\Sneaker_Releases.xlsx', usecols=['Brand', 'Model'])
```

```python
from itertools import chain
```

```python
sneakers = list(map(str.lower, set(chain(*sneakers.values))))
```

```python
data = data[['caption', 'text']]
```

```python
data = data.dropna()
```

```python
for i, c, t in data.itertuples():
    print(c, i)
    print('\n')
```

```python
found_list = []
for sneaker in sneakers:
    for i, c, t in data.itertuples():
        if sneaker in c.lower():
            found_list.append(i)
```
